### Methods

A method is a function that is only available to a specific data type. 

How we call str 
methods:

**Examples of String Methods:**
1. .replace()
2. .lower()
3. .upper()

```python
colour_statement = "Isaac's favourite colour is blue."
print(colour_statement)
```

```python
right_colour_statement = colour_statement.replace("blue", "brown")
print(right_colour_statement)
```

#### Changing Cases

```python
current_top_f1_player ="Max Verstappen"

# Convert this tolowercase

current_top_f1_player = current_top_f1_player.lower()
print(current_top_f1_player)
```

```python
# Change it to uppercase

current_top_f1_player = current_top_f1_player.upper()
print(current_top_f1_player)
```

Data Structures in Python

Lists in Python

- Lists store multiple values in a single variable
  - Lists can contaion any combuination of data type eg boolean, string, int

- Lists are ordered
  - One can use subsetting or indexing

```python
fuel_prices = [90, 150, 100, 194, 60, 76]
print(fuel_prices)
```

#### Accessing Elements of a list
- We use the

```python
print(fuel_prices)
```

```python
(fuel_prices)[2]
```

Slicing in Lists / Accessing Multiple Elements in a List

```python
fuel_prices[2:5]
```

```python
# Get the first three fuel prices
# These are values from index zero to index two (not inclusive of three)
fuel_prices[:3]
```

```python
# Access all the elements from the fourth index onwards

fuel_prices[3:]
```

##### Alternating Access

```python
print(fuel_prices)
```

```python
# Accessing Every other element in the List
fuel_prices[::2]
```

```python
# Access every 3rd element starting from the second index element(index 1)
fuel_prices[1::3]
```
